```python
interest_tokens = interests_str.split("\n")
interest_dict = {}
keys = []
key_count = -1 
for i in interest_tokens:
    if ":" in i:
        keys.append(i.replace(":","").strip())
        key_count += 1
    else:
        try:
            interest_dict[keys[key_count]]
            interest_dict[keys[key_count]].append(i.strip())
        except KeyError:
            interest_dict[keys[key_count]] = [i.strip()]
#         interest_dict[i.replace(":","").strip()]
interest_dict
```

```python
archetypes = ['Places', 'World Affairs', 'Sports', 'Languages', 'Hustle',
       'Knowledge', 'Wellness', 'Life', 'Tech', 'Entertainment',
       'Identity', 'Faith', 'Arts']
archetype_emojis = ["🗺️", "🌎","🏆",  "💬","🔥", "💡", "🌻", "🤔", "💻","🥳", "📛","🎨", "🛐"]
archetypes_and_emotes_dict = dict(zip(archetypes, archetype_emojis))
data_dict = {}
data_dict["archetype"] = []
data_dict["archetypeEmoji"] = []
data_dict["name"] = []

for k,v in interest_dict.items():
    print(k,v)
    data_dict["archetype"].extend([k]*len(v))
    data_dict["archetypeEmoji"].extend([archetypes_and_emotes_dict[k]]*len(v))
#     data_dict["archetypeEmoji"].extend([k])
    data_dict["name"].extend(v)
df = pd.DataFrame(data_dict)
# df
df.to_csv("interests.csv", index=False)
df["archetype"].unique()
```
